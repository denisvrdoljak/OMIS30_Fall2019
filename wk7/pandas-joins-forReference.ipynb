## Pandas Groupby

When looking at data it is good to be able to
- Count all the values under a category
- Sum all the values under a category
- Do other operations by category (mean, standard deviation, other operations)

For these operations pandas has a groupby function that groups that data on one (or more) columns and then can aggregate that data.  Some example on why you would want to use this: 
- Sum all the customer purchases made in different stores
- Find the average fuel efficiency for different cars
- Count the number of times a visitor came to different website

```python
# The below cell just writes a simple datafile so we can show the groupby
# The %%writefile is a magic command for jupyter notebooks - to write that cell to the filename
```

```python
%%writefile group.csv
Index,State,Amount
0,CA,100
1,CA,205
2,CO,367
3,NY,365
4,GA,525
5,CA,150
6,CO,342
```

```python
# Here we are reading in the file we just created above
# And then grouping by the state and summing the amount

df_g = pd.read_csv('group.csv')
df_g.groupby(['State']).sum()
```

```python
# Grouping and showing the count ( or # of entries in which that state appears)

df_g.groupby(['State']).count()
```

```python
# Grouping and showing the mean ( or # of entries in which that state appears)

df_g.groupby(['State']).mean()
```

## Dataframe joins

If you have two different files you might want to join them up to make one larger file.  Pandas has a few ways to do this:
- df.append() : takes two dataframes with the same columns and adds the second one to the bottom of the first
- Append is useful when you have two dataframes that are divided by time (for example months).  You can merge the January & February smaller dataframes into a large one.



- pd.concat() : takes two dataframes with the same rows and adds the second one to the right of the first
- Concat is useful when you have two dataframes that are ordered by the same value (maybe an ID or customer #) and have the same number of rows (for example one dataframe contains address information while the other has phone #s).  It is important that both dataframes have the same number of rows or when combined the data will not match up correctly.



- merge() : Takes two dataframes and joins them on a key that is the same for each dataframe.  The column containing the ID has to have the same column name.
- merge/joins are useful when you have two dataframes that do not have the same number of rows.  For example, customer information in one dataframe and if a customer is signed up for a certain program in another dataframe.  Not all customers might be signed up for the program so their IDs might not be in the second dataframe.
- Pandas has 4 types of merge/joins:  inner, outer, left and right.  



The syntax of each is demonstrated below:

```python
df1.append(df1_1)
```

```python
pd.concat([df1,df2], axis=1)
```

```python
pd.merge(df1, df2_2, on='ID', how = 'outer')
```

```python
pd.merge(df1, df2_2, on='ID', how = 'inner')
```

```python
pd.merge(df1, df2_2, on='ID', how = 'left')
```

```python
pd.merge(df1, df2_2, on='ID', how = 'right')
```

## Pandas Charts

Graphs/charts can be done in pandas using a variety of charting functions.  Pandas uses the Python library matplotlib to produce the charts.  A full guide can be found here:

- https://pandas.pydata.org/pandas-docs/stable/visualization.html
- Chart types here: https://pandas.pydata.org/pandas-docs/stable/visualization.html#other-plots



```python
# Review on a graphing example (making a simple histogram from a column)

df['PERCENT US CITIZEN'].hist(bins = 25)
```

```python
# To make labels for each axis:
# Need to modify the 'graph object' that is returned from histogram call
# And then take that object and 'set' the labels
# Finally plt.plot() shows the chart

ax = df['PERCENT US CITIZEN'].hist(bins = 25)
ax.set(xlabel='Percent US Citizen', ylabel='Total Count', title = "Histogram of US Citizenship Percentage")
plt.plot()
```

```python
# This example shows how to change the color by using arguments to the histogram plotting function
# The full color codes options: ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

ax = df['PERCENT US CITIZEN'].hist(bins = 25, color = 'r', alpha = 0.5)
ax.set(xlabel='Percent US Citizen', ylabel='Total Count', title = "Histogram of US Citizenship Percentage")
plt.plot()
```

##### Time-series chart

A time-series line chart can be built using pandas the same way as the other charts. In pandas, you do need to specifically make a column of type datetime to be able to use that time column (otherwise pandas might just consider it a string for sorting/calculations). Below is an example using a small dataset on how to build a time series chart

```python
%%writefile time_series.csv
Month,Snow,Rain,Sun
2018-01,15,0,16
2018-02,10,0,18
2018-03,3,6,22
2018-04,1,5,24
2018-05,0,10,21
```

```python
# Reading in the toy dataset we made above

ts = pd.read_csv("time_series.csv")
```

```python
# Checking the type of the column month - it is an object rather than a datetime type so we need to fix it.

ts.info()
```

```python
# Converting the object to a datetime type
# Full conversion syntax is at: https://docs.python.org/3.6/library/datetime.html#strftime-and-strptime-behavior

ts.Month = pd.to_datetime(ts.Month, format="%Y-%m")
```

```python
# Re-checking the types - confirmed month is now a datetime type

ts.info()
```

```python
# We can plot multiple columns over the time period.  
# Notice that the months dont have a label and even though we only have 5 entries there are 9 ticks...

ax = ts.plot(x="Month", y=["Snow", "Rain","Sun"])
ax.set(xlabel='Month', ylabel='Days', title="# of Snow Days per Month")
plt.plot()
```

```python
# To fix the x-axis tick issue when need to specifically tell pandas/matplotlib how many ticks we want
# It needs to be a range - its easiest to make it the range from 0 to the length of our dataframe
# This allows us to flexible if our dataframe changes length - as the ticks will change with it


ax = ts.plot(x="Month", y=["Snow", "Rain","Sun"], xticks=range(0, len(ts)))
ax.set(xlabel='Month', ylabel='Days', title="# of Snow Days per Month")
plt.plot()
```
